```python
from seismic.db.corr_hdf5 import CorrelationDataBase
path = f'../xstations_no_response_removal_5_1.0-2.0_wl55.0_1b_SW/UW-UW.EDM-HSR.h5'

with CorrelationDataBase(path, mode='r') as cdb:
    # cst = cdb.get_data('UW-UW', 'HSR-HSR', 'EHZ-EHZ', 'subdivision')
    cst = cdb.get_data('UW-UW', 'EDM-HSR', 'EHZ-EHZ', 'stack_86398')
cst.plot(type='heatmap', cmap='seismic')
```

```python
# mute middle
for ctr in cst:
    npts = ctr.stats.npts
    ctr.data[int(npts*2.2/5):int(npts*2.8/5)] = 0
```
